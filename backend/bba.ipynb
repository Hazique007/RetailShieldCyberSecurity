```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
import os
import joblib
```

```python
data = pd.read_csv("DSL-StrongPasswordData.csv")
```

```python
cmu = data.copy()
```

```python
cmu.info()
```

```python
feature_col = cmu.columns[3:]
users = sorted(cmu['subject'].unique())
```

```python
feature_col
```

```python
users
```

```python
# feature_cols = cmu.columns[3:]
print(f"ðŸ§  Using {len(feature_col)} keystroke features.")
```

```python
X = cmu[feature_col].values
```

```python
print("ðŸ“Š Scaling features...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print("ðŸ§  Training One-Class SVM...")
model = OneClassSVM(kernel='rbf', gamma='auto', nu=0.05)
model.fit(X_scaled)
```

```python
MODEL_DIR = "global_model"
```

```python
os.makedirs(MODEL_DIR, exist_ok=True)
model_path = os.path.join(MODEL_DIR, "oneclass_svm.pkl")
scaler_path = os.path.join(MODEL_DIR, "scaler.pkl")

joblib.dump(model, model_path)
joblib.dump(scaler, scaler_path)

print(f"âœ… Model saved to: {model_path}")
print(f"âœ… Scaler saved to: {scaler_path}")
```
